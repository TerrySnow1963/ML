```python
fix, axes = plt.subplots(10,12, figsize=(12,13),subplot_kw={'xticks':(),'yticks':()})
for i, (component, ax) in enumerate(zip(imfiles, axes.ravel())):
    if i < 119:
        im = mpimg.imread(imfiles[i],'grey')
        ax.imshow(im,cmap='gray')
        ax.set_title(imfiles[i][-8:-5], fontsize=10)    
```
